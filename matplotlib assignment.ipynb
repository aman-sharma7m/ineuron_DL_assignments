```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as pl
%matplotlib inline
```

```python
ds=sns.load_dataset('titanic')
ds.head()
```

```python
ds.groupby('sex').size().plot(kind='pie',autopct='%.2f%%')
```

```python
sns.scatterplot(x=ds['fare'],y=ds['age'],hue=ds['sex'])
pl.show()
```

```python
max1=[39,41,43,47,49,51,45,38,37,29,27,25]
y=[i for i in range(0,12)]
min1=[21,23,27,28,32,35,31,28,21,19,17,18]
pl.figure(figsize=(20,10))
pl.xlabel('months')
pl.ylabel('temp(C)')
pl.scatter(y=max1,x=y,color='r')
pl.plot(y,max1,'r')
pl.scatter(y=min1,x=y)
pl.plot(y,min1)
```
